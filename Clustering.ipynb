```python
# scaling the features

from sklearn.preprocessing import StandardScaler

s_obj = StandardScaler()
scaled_data = pd.DataFrame(s_obj.fit_transform(data.drop("country", 1)))
scaled_data.columns = data.drop("country", 1).columns
```

# Hierarical Clustering

Single Linkage

```python
# Hierarical clustering - Single Linkage

from scipy.cluster.hierarchy import linkage, dendrogram,cophenet, cut_tree
from scipy.spatial.distance import pdist
plt.figure(figsize=[10,10])
merg_single = linkage(scaled_data, method='single')
dendrogram(merg_single, leaf_rotation=90)
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distances')
plt.show()
```

**dendogram doesn't seeem matching with the businees logic, as majority of the data is kept in a 
single cluster(even if i chose 5 clusters )**

Complete Linkge

```python
# Hierarical clustering - Complete Linkage

from scipy.cluster.hierarchy import linkage, dendrogram,cophenet, cut_tree
from scipy.spatial.distance import pdist
plt.figure(figsize=[10,10])
merg_complete = linkage(scaled_data, method='complete')
dendrogram(merg_complete, leaf_rotation=90)
plt.title('Dendrogram')
plt.xlabel('Data Points')
plt.ylabel('Euclidean Distances')
plt.show()
```

**complete method seems better than single linkage method. But still this is majorly making two cluster. moving forward with complete linkage between single/complete linkage methods**

```python
# Above both dendogram doesn't seems to be very useful. But to compare with kmean I am considering 
# 3 as cluster size(seeing dendrogram)

H_cluster_labels = cut_tree(merg_complete, n_clusters=3).reshape(-1, )
H_cluster_labels
```

```python
data_H = data.copy()

data_H["cluster_id"] = H_cluster_labels
```

# K-Means

```python

from sklearn.cluster import KMeans

SSD = []

for k in range(1,10):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaled_data)
    SSD.append(kmeans.inertia_)
    
# To make elbow curve
plt.plot(range(1,10),SSD)
plt.ylabel("SSD")
plt.xlabel("Clusters");
# It seems like Number 4 cluster seems good choice from the below graph. 
```

```python
data.drop("country",axis = 1)
```

```python
scaled_data.shape
```

```python
# silhouette analysis
from sklearn.metrics import silhouette_score
n_cluster = [2, 3, 4, 5, 6, 7, 8]



for i in n_cluster:
    k_mean = KMeans(n_clusters=i)
    k_mean.fit(scaled_data)
    labels = k_mean.labels_
    silhouette_num = silhouette_score(scaled_data, labels)
    print(i, silhouette_num)
    
    
```

from silhoutter score , it seems 5 cluster are the optimum number of cluster

```python
# model will be with 5 clusters:
k_mean = KMeans(n_clusters=5)
k_mean.fit(scaled_data)
labels = k_mean.labels_
```

```python
pd.Series(labels).value_counts()
```

It seems like 5 clusters are best, from silhoutter score. 
But I would prefer going with 3, as 2 of the cluster has only 3+3 observation

```python
# Final model will be with 3 clusters:
k_mean = KMeans(n_clusters=3)
k_mean.fit(scaled_data)
labels = k_mean.labels_
```

```python
data["cluster_id"] = k_mean.labels_
data
```
